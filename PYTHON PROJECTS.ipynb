```python
#Create Theatre ticket booking system

print("="*15,"Welcome to theatre ticket booking system","="*15)
print()

print("Normal Seat"," "*15,"Recliner Seat"," "*15,"Popcorn Combo")
print("250 Rupees"," "*17,"450 Rupees"," "*18,"200 Rupees")

print()
print("-"*73)


total = 150
normalseat = 140
reclinerseat = 10

popcorn = 200

nprice = 250 
rprice = 450

while True:
    name = input("Customer name: ")
    seats = int(input("How many seats do you prefer?: "))
    
    choice = input("Which seats do you prefer (Normal/Recliner)?: ").lower()
    
    if choice == "normal":
        combo = input("Would you like to add popcorn (yes/no)? ")
        if combo == "yes":
            print("-"*40)
            print(name ,"Your total amount with popcorn is: ", (seats*nprice)+popcorn)
            print("Thank You, Visit again!")
            print("-"*40)
            normalseat -= seats
            if normalseat > 0:
                print(normalseat, " seats left")
            else:
                print("Housefull")
        else:
            print("-"*40)
            print(name ,"Your total amount is ", seats*nprice)
            print("Thank You, Visit again!")
            print("-"*40)
            normalseat -= seats
            if normalseat > 0:
                print(normalseat, " seats left")
            else:
                print("Housefull")
    else:
        combo = input("Would you like to add popcorn (yes/no)? ")
        if combo == "yes":
            print("-"*40)
            print(name, "Your total amount with popcorn is ",  (seats*rprice)+popcorn)
            print("Thank You, Visit again!")
            print("-"*40)
            reclinerseat -= seats
            if reclinerseat > 0:
                print(reclinerseat, " seats left")
            else:
                print("housefull")
        else:
            print("-"*40)
            print(name, "Your total amount is ", seats*nprice)
            print("Thank You, Visit again!")
            print("-"*40)
            reclinerseat -= seats
            if reclinerseat > 0:
                print(reclinerseat, " seats left")
            else:
                print("housefull")
            
            
    repeat = input("Next person(yes/no)?: ")
    if repeat == "no":
        break
        
        
print("Counter closed") 
```
